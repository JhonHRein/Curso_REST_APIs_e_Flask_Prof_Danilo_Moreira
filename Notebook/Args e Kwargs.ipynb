```python
# Exemplo de Args. 
def listar_frutas(*args):
    print("=== Frutas recebidas ===")
    for fruta in args:
        print("-", fruta)
```

```python
def listar_frutas(*args):
    print("=== Frutas recebidas ===")
    for fruta in args:
        print("-", fruta)

quantidade = int(input("Quantas frutas você quer informar? "))

frutas = []

for indice in range(quantidade):
    nome_fruta = input(f"Digite o nome da fruta {indice + 1}: ")
    frutas.append(nome_fruta)
```

```python
# Agora chamamos a função usando o *
listar_frutas(*frutas)
```

```python
#Exemplo de **KWARGS 
#Criando Funçao 

def ficha_aluno(**kwargs):
    print("=== Ficha do aluno ===")
    for chave, valor in kwargs.items():
        print(f"{chave}: {valor}")
```

```python
quantidade_info = int(input("Quantas informações você quer colocar na ficha? "))

ficha = {}

for indice in range(quantidade_info):
    chave = input(f"Nome da informação {indice + 1} (ex: nome, idade, cidade): ")
    valor = input("Valor dessa informação: ")
    ficha[chave] = valor

ficha_aluno(**ficha)
```
