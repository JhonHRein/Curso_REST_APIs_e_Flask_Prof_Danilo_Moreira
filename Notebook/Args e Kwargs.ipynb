```python
def listar_frutas(*args):
    print("=== Frutas recebidas ===")
    for fruta in args:
        print("-", fruta)

quantidade = int(input("Quantas frutas você quer informar? "))

frutas = []

for indice in range(quantidade):
    nome_fruta = input(f"Digite o nome da fruta {indice + 1}: ")
    frutas.append(nome_fruta)
```

```python
# Agora chamamos a função usando o *
listar_frutas(*frutas)
```
